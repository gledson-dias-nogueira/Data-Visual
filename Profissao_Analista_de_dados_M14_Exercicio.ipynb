## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head(5940)
```

```
# gráfico do exercício 1

diamonds = round(data[["cut", "price"]].groupby("cut").agg("mean"),2)
```

```
grafico = sns.barplot(data=diamonds, x="cut", y="price", ci=None, palette="pastel")
grafico.set(title='Média valor diamentes por modelo', xlabel='Modelos de cortes ', ylabel='Valor médio');
```

**Insight do gráfico 1**: Apesar do "Fair" ser uma cetegoria inferior o valor médio pago nele foi quase igual ao "Premium" enquanto o "ideal" é o mais ecônomico

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = data[data["month"]=="Dec"]

grafico = sns.lineplot(data=flights, x="year", y="passengers", palette="pastel")
grafico.set(title='Passageiros por ano', xlabel='Ano', ylabel='Passageiros');
```

**Insight do gráfico 2**: Veio em uma forte tenência de crescimento até o periode de 56 a 58 quando estagnou e logo após retomou seu forte crescimento

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

plt.figure(figsize=(10, 6))


grafico = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")
grafico.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros');
grafico.get_legend().set_title("Ano");


plt.ylim(80, 700)
plt.legend(loc='upper left', bbox_to_anchor=(0.01, 1), fontsize='small')
```

**Insight do gráfico 3**: Nos 10 anos avliados se repetiu uma tendencia bem clara na proporção de passagerios por cada mês do ano
